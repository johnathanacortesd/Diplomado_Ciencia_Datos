```python
# Cuaderno 2
# Importar las bibliotecas necesarias
import pandas as pd
from IPython.display import display

# Punto 3: Clase de Estadísticas
class DescriptiveStats:
    def __init__(self, file_path):
        self.df = pd.read_csv(file_path)

    def basic_stats(self):
        return self.df.describe()

    def variable_names(self):
        return self.df.columns.tolist()

    def variable_summary(self, variable_name):
        if variable_name in self.df.columns:
            variable_data = self.df[variable_name]
            stats = {
                "Mean": variable_data.mean(),
                "Median": variable_data.median(),
                "Standard Deviation": variable_data.std(),
                "25th Percentile": variable_data.quantile(0.25),
                "50th Percentile": variable_data.quantile(0.50),
                "75th Percentile": variable_data.quantile(0.75)
            }
            display(pd.DataFrame(stats, columns=["Statistics"]))
            return stats
        else:
            return "Variable not found."

# Prueba de la Clase de Estadísticas
stats_obj = DescriptiveStats("iris.csv")
print(stats_obj.basic_stats())
print(stats_obj.variable_names())
print(stats_obj.variable_summary("sepal.length"))
```
