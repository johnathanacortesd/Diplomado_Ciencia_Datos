##statsmodels
##¿Qué es statsmodels?

Statsmodels es una biblioteca de Python diseñada para realizar análisis estadísticos y modelado de datos. Se enfoca en la estimación y pruebas de modelos estadísticos, como regresión lineal, regresión logística, análisis de series temporales y más.

##Ejemplos de uso en la vida real:

Análisis de regresión lineal: Statsmodels es útil para analizar la relación entre una variable dependiente y una o más variables independientes. Por ejemplo, podrías utilizarlo para estudiar cómo las horas de estudio influyen en las calificaciones de los estudiantes.

Modelado de series temporales: Statsmodels es adecuado para analizar datos de series temporales, como las ventas mensuales de una tienda minorista, para identificar tendencias, estacionalidades y hacer predicciones futuras.

Pruebas de hipótesis: Puedes utilizar statsmodels para realizar pruebas de hipótesis, como pruebas t y pruebas F, para evaluar la significancia de coeficientes en un modelo y determinar si ciertas variables tienen un impacto significativo.

Regresión logística: Statsmodels es útil para modelar problemas de clasificación binaria o multinomial, como predecir si un paciente tiene una enfermedad en función de múltiples variables médicas.

Análisis de varianza (ANOVA): Puedes usar statsmodels para realizar análisis de varianza para comparar las medias de múltiples grupos y determinar si existen diferencias significativas entre ellos.

Estos ejemplos proporcionan una visión general de cómo se pueden utilizar scikit-learn y statsmodels en aplicaciones del mundo real. Ambas bibliotecas son extremadamente versátiles y pueden aplicarse a una amplia variedad de problemas de análisis de datos y aprendizaje automático.

## 3. Exploración de scikit-learn (sklearn)
Realicemos ejercicios prácticos utilizando scikit-learn:

a) Importar y cargar conjuntos de datos:

```python
from sklearn.datasets import load_diabetes
import pandas as pd

# Cargar el conjunto de datos "diabetes"
data = load_diabetes()

# Obtener características (X) y variable objetivo (y)
X = data.data
y = data.target

# Convertir a un DataFrame de pandas para una mejor visualización
df = pd.DataFrame(X, columns=data.feature_names)
df['TARGET'] = y

# Visualizar las primeras filas del conjunto de datos
print(df.head())
```

##Preprocesamiento de datos
Implementar algoritmos de aprendizaje supervisado

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

##Implementar algoritmos de aprendizaje supervisado

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)
```

##Evaluar y validar modelos
Evalúa y valida tus modelos utilizando métricas apropiadas como el coeficiente de determinación (R^2) y el error cuadrático medio (MSE).

```python
from sklearn.metrics import r2_score, mean_squared_error

y_pred = model.predict(X_test)
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

print(f'R^2: {r2}')
print(f'MSE: {mse}')
```

Aproximadamente el 45.26% de la variabilidad en tus datos de diabetes es explicada por el modelo de regresión lineal.

El valor del MSE es 2900.19, lo que significa que, en promedio, el modelo comete un error de aproximadamente 2900.19 unidades al predecir los valores de diabetes.

## Interpretar resultados estadísticos y realizar pruebas de hipótesis:
La interpretación de resultados estadísticos es fundamental en el análisis de datos. En el contexto de la regresión lineal, algunos de los elementos clave a considerar en el resumen de resultados de statsmodels son:

##Coeficientes (coef):
Representan las relaciones entre las variables independientes y la variable dependiente. Un coeficiente positivo significa que un aumento en la variable independiente está asociado con un aumento en la variable dependiente, y viceversa.

##Valor p (p-value):
Es importante para las pruebas de hipótesis. Un valor p bajo (generalmente menor que 0.05) indica que la variable independiente es estadísticamente significativa en la predicción de la variable dependiente.

##R-cuadrado (R²):
El coeficiente de determinación R² indica la cantidad de varianza en la variable dependiente que es explicada por las variables independientes. Un valor cercano a 1 indica un buen ajuste del modelo a los datos.

##Intervalos de confianza (0.025 y 0.975):
Estos intervalos indican el rango dentro del cual es probable que se encuentre el coeficiente con cierto nivel de confianza.

Ejemplo de código para interpretar resultados estadísticos y realizar pruebas de hipótesis en una regresión lineal con statsmodels:

```python
import statsmodels.api as sm

# Realizar una regresión lineal
X_const = sm.add_constant(X_scaled)
model = sm.OLS(y, X_const).fit()

# Imprimir un resumen de los resultados
print(model.summary())

# Ejemplo de prueba de significancia de un coeficiente
coef_intercepto = model.params[0]  # El índice 0 corresponde al coeficiente del intercepto
p_valor_intercepto = model.pvalues[0]  # El índice 0 corresponde al p-valor del intercepto

if p_valor_intercepto < 0.05:
    print(f"El coeficiente de intercepto ({coef_intercepto}) es significativo.")
else:
    print(f"El coeficiente de intercepto ({coef_intercepto}) no es significativo.")
```

## Realizar pruebas de hipótesis:
Realizar pruebas de hipótesis es una parte esencial del análisis estadístico. En el contexto de modelos de regresión, las pruebas de hipótesis se utilizan para evaluar la significancia de los coeficientes y determinar si las variables independientes tienen un efecto estadísticamente significativo en la variable dependiente. En este ejemplo, mostraré cómo realizar una prueba de hipótesis para evaluar la significancia de un coeficiente utilizando la biblioteca statsmodels.

Se usarán las columnas numéricas como índices:

```python
# Paso 1: Importar las bibliotecas necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm

# Paso 2: Cargar el conjunto de datos de diabetes
diabetes = load_diabetes()
X = diabetes.data  # Variables independientes
y = diabetes.target  # Variable dependiente

# Crear un DataFrame para facilitar la manipulación de datos (opcional)
df = pd.DataFrame(data=np.c_[diabetes.data, diabetes.target], columns=diabetes.feature_names + ['target'])

# Paso 3: Ajustar un modelo de regresión lineal utilizando scikit-learn
model = LinearRegression()
model.fit(X, y)

# Imprimir los coeficientes y el intercepto del modelo
print("Coeficientes:", model.coef_)
print("Intercepto:", model.intercept_)

# Paso 4: Ajustar un modelo de regresión lineal utilizando statsmodels
X_with_intercept = sm.add_constant(X)  # Agregar una columna de intercepto
model_sm = sm.OLS(y, X_with_intercept).fit()

# Imprimir un resumen del modelo ajustado
print(model_sm.summary())

# Paso 5: Realizar pruebas de hipótesis (ejemplo de prueba de significancia del coeficiente de 'bmi')
coef_bmi = model_sm.params[2]  # El índice 2 representa el coeficiente de 'bmi'
p_value_bmi = model_sm.pvalues[2]

if p_value_bmi < 0.05:
    print(f"El coeficiente de 'bmi' ({coef_bmi}) es significativo.")
else:
    print(f"El coeficiente de 'bmi' ({coef_bmi}) no es significativo.")

# Paso 6: Crear gráficos y visualizaciones relevantes (ejemplo de gráfico de dispersión)
plt.scatter(df['bmi'], df['target'], c='blue', label='Datos reales')
plt.plot(df['bmi'], model.predict(X), c='red', label='Regresión lineal')
plt.xlabel('BMI')
plt.ylabel('Target')
plt.legend()
plt.title('Gráfico de dispersión: BMI vs. Target')
plt.show()
```

##Crear gráficos y visualizaciones
Statsmodels también te permite crear gráficos y visualizaciones relevantes para el análisis de regresión logística, como gráficos ROC y curvas de calibración. Aquí hay un ejemplo de cómo crear una curva ROC:

```python
# Importar las bibliotecas necesarias
import matplotlib.pyplot as plt

# Gráfico de dispersión: BMI vs. Target
plt.scatter(df['bmi'], df['target'], c='blue', label='Datos reales')
plt.plot(df['bmi'], model.predict(X), c='red', label='Regresión lineal')
plt.xlabel('BMI')
plt.ylabel('Target')
plt.legend()
plt.title('Gráfico de dispersión: BMI vs. Target')
plt.show()

# Histograma de los residuos
residuos = y - model.predict(X)
plt.hist(residuos, bins=20, color='green', alpha=0.5)
plt.xlabel('Residuos')
plt.ylabel('Frecuencia')
plt.title('Histograma de Residuos')
plt.show()

```
